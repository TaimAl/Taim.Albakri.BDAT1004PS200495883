```python
# Question 10

def frequencies(word):
    alpha = {'a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z'} #intrdoucing a dictionary 
    for n in word: 
        if n in alpha: 
            alpha[n] += 1
        else:
            alpha[n] = 1
    return alpha #counting the frequesncies 


frequencies ('apple')
```
